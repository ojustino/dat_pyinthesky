```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(74)
```

```python
R = 4
N = int(1e4)
```

```python
xs = R * (np.random.rand(N) * 2 - 1)
ys = R * (np.random.rand(N) * 2 - 1)
valid = xs**2 + ys**2 <= R**2
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(xs[valid], ys[valid], s=1, c='#e56020')
```

```python
4 * valid.sum() / N
```
